# Logistic map
Logistic map is a polynomial mapping of degree 2, which can simulate the chaotic behaviour that can arise from very simple non-linear dynamical equations. Mathematically, the general form of the logistic map is written as:
$$
x_{n+1} = a x_{n}{1-x_n}
$$
where, $x_n$ is a number between 0 and 1, that represents the ratio of existing value to the maximum possible value. Similarly the sequence $x_0, x_1, x_2$ is called trajectory or orbit.

In part a),  over the interval of n = 0,---,200, the logistic map was iterated using general approach for the different values of r. For different r values the graph of x as a function of n were plotted which agreed with the theoritical plot from the lecture note. Similarly, for part b) the iterations were computed for the maps of $ f(x) = x.exp(4r(1-x)) $ and $f(x) = rsin(\pi x)$ and resulting plots were obtained for the same parameters as in part a).

```python


import numpy as np
import matplotlib.pyplot as plt

# Given Parameters
r = [0.2, 0.4, 0.73, 0.88]
x0 = [0.8, 0.8, 0.1,0.09]
N = 200



#Generate x, y, z for each value of n

for i in range(len(r)):
    x = x0[i] + np.zeros(N)
    y = x0[i] + np.zeros(N)
    z = x0[i] + np.zeros(N)
    a= 4*r[i] 
    
    # Iterating the logistic map
    for n in range(N-1):
        x[n+1] = a*x[n]*(1-x[n])
        y[n+1] = y[n]*np.exp(4*r[i]*(1-y[n]))
        z[n+1] = r[i]*np.sin(np.pi*z[n])
       
        
  
    plt.plot(x,'-', label = 'r={}, x0={}, $f(x_n) = ax_n(1-x_n)$a'.format(r[i],x0[i]))
    plt.xlabel('n')
    plt.ylabel('x')
    plt.legend()
    plt.show()
    
    plt.plot(y,'-', label = 'r={}, x0={}, $f(x) = x exp(4r(1-x))$'.format(r[i],x0[i]))
    plt.ylabel('n')
    plt.ylabel('x')
    plt.legend()
    plt.show()
 
    
    plt.plot(z,'-', label = 'r={}, x0={}, f(x) = r sin($\pi$ x)'.format(r[i],x0[i]))
    plt.xlabel('n')
    plt.ylabel('x')
    plt.legend()
    plt.show()
    
    
    
 
```

Part c) For the given values of r = 0.92, $x_0 = 0.6$ and $\Delta x_0$= $10^{-5}$,  ln$\left|\frac{\Delta x_n}{\Delta x_0}\right|$ was plotted as a function of n.

```python
import numpy as np
import matplotlib.pyplot as plt

r =0.92
a= 4*r 
x10 = 0.6
dx0 = 10**(-5)
x20 = x10 + dx0
N = 200
nl =np.linspace(0, 200, num=200)

#Generate x1, x2, x3 for each value of r
x1 = x10 + np.zeros(N)
x2, x3 = x1, x1
y1 = x20 + np.zeros(N)
y2, y3 = y1, y1

#iterate system for N-1 times
for n in range(N-1):
    x1[n+1] = a*x1[n]*(1-x1[n])
    y1[n+1] = a*y1[n]*(1-y1[n])
    
    x2[n+1] = x2[n]*np.exp(4*r*(1-x2[n]))
    y2[n+1] = y2[n]*np.exp(4*r*(1-y2[n]))
    
    x3[n+1] = r*np.sin(np.pi*x3[n])
    y3[n+1] = r*np.sin(np.pi*y3[n])
    

dx1 = dx0 + np.zeros(N)

Delx = []
for i in range(N):
    dx1[i] = y1[i] - x1[i]
    c = np.log(abs(dx1[i])/dx0)
    Delx.append(c)

plt.plot( nl,Delx, 'k.-', label = '$r=0.92, x_0 = 0.6, \Delta x_0 = 10^{-5}$')    
plt.ylabel(r'ln$\left|\frac{\Delta x_n}{\Delta x_0}\right|$', size=16)
plt.xlabel('n')
plt.legend()
plt.show()
```

# Lyapunov eponents

Lyapunov exponents is used to check the sensitivity to initial conditions in 1-D maps required of a chaotic system. For some initial condition $x_0$, and $\delta_n$ being the seperation after n iterates.
Now, if $\left|\delta_n\right|\approx \left|\delta_0\right|exp(n\lambda)$, then $\lambda$ is called the Lyapunov exponent. More presicely, the Lyapunov exponent for the orbit starting at $x_0$ is defined to be:
$$
\lambda = \lim_{n \to \infty} \left(\frac{1}{n} \sum_{i=0}^{n-1} ln{f'(x_i)}\right)
$$
For stable fixed points and cycles $\lambda$ is negative whereas it is positive for chaotic attractors.

Similarly, for 3 different maps the plot of $\lambda$ as a function of r was plotted where the Lyapunov eponent were calulated to be 0.6939, 1.963 and 0.696 for the general logistic map, f(x) = xexp(4r(1.x)) and f(x) = $rsin(\pi x)$ and the super-stable state were obtained at r value of 0.5, 0.35 and 0.55 respectively. 



```python

import numpy as np
import matplotlib.pyplot as plt



s1,s2,s3 = [],[], []

rarr = np.linspace(0,1,200)

#Generate x1, x2, x3 for each value of r
for r in rarr:
    
    a= 4*r 
    x10 = 0.6
    dx0 = 10**(-5)
    x20 = x10 + dx0
    N = 200
    nl =np.linspace(0, 200, num=200)
    
    x1 = x10 + np.zeros(N)
    x2, x3 = x1, x1
    y1 = x20 + np.zeros(N)
    y2, y3 = y1, y1
    
    #iterate system for N-1 times
    for n in range(N-1):
        x1[n+1] = a*x1[n]*(1-x1[n])
        y1[n+1] = a*y1[n]*(1-y1[n])
        
        x2[n+1] = x2[n]*np.exp(4*r*(1-x2[n]))
        y2[n+1] = y2[n]*np.exp(4*r*(1-y2[n]))
        
        x3[n+1] = r*np.sin(np.pi*x3[n])
        y3[n+1] = r*np.sin(np.pi*y3[n])
        
   
        
   
    dx1 = dx0 + np.zeros(N)
    dx2, dx3 = dx1,dx1
    Delx1, Delx2, Delx3 = [dx0], [dx0], [dx0]
    
    for i in range(1, N):
        dx1[i] = (y1[i] - x1[i])
        dx2[i] = (y2[i] - x2[i])
        dx3[i] = (y3[i] - x3[i])
        
        c1 = np.log(abs(4*r*(1-2*x1[i])))
        c2 = np.log(abs((1-4*r*x2[i]*np.exp(4*r*(1-x2[i])))))
        c3 = np.log(abs(np.pi*r*(np.cos(np.pi*x3[i]))))
        
       
        Delx1.append(c1)
        Delx2.append(c2)
        Delx3.append(c3)
        
     #Calculating Lyapunov exponents
    lamda1 = np.sum(Delx1)/N
    lamda2 = np.sum(Delx2)/N
    lamda3 = np.sum(Delx3)/N
    s1.append(lamda1) 
    s2.append(lamda2)
    s3.append(lamda3)
print('Lambda_1 =', lamda1)
print('Lambda_2 =',lamda2)
print('Lambda_3 =',lamda3)

np.seterr(divide = 'ignore') 
plt.plot(rarr, s1,'k.-', label ='$f(x_n) = ax_n(1-x_n)$')  
plt.xlabel('r') 
plt.ylabel('$\lambda$') 
plt.legend()
plt.show()

plt.plot(rarr, s2,'k.-', label ='$f(x) = x exp(4r(1-x))$')  
plt.xlabel('r') 
plt.ylabel('$\lambda$') 
plt.legend()
plt.show()

plt.plot(rarr, s3,'k.-', label = 'f(x) = r sin($\pi$ x)')  
plt.xlabel('r') 
plt.ylabel('$\lambda$') 
plt.legend()
plt.show()

    
```
