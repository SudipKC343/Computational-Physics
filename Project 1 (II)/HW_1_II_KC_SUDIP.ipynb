## Linear congruential random number generator

In the first part of the experiment, a random number generator is created using linear congruential method:
$$
x_{i+1} = (ax_i + b) \quad mod \quad  c
$$
where a, b and c are the integer parameters, termend "multiplier", "increment" and "modulus" respectively and the "mod" specifies the modulus operator. For the given values of (a, b, c , $x_0$) = (57, 1, 256, 10) random numbers are generated with the period of 256 and the pairs ($x_i$, $x_{i+1}$) are plotted on a 2D scatterplot. In the first plot where 500 different random numbers were plotted, an autocorrelation pattern is seen on the plot of the pairs ($x_i$, $x_{i+1}$) due to the repeating random numbers after the period. In the second plot  taken for only one period, the random numbers are seen to be non-uniformly distributed. 


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

a = 57     #multiplier term
b = 1    #increment term
c = 256    #modulus
x0 = 10     #seed value


#Function to generate random numbers
def LCS(x0,a,b,c,randomnums,noofrandomnums):
    randomnums[0] = x0
    for i in range(1, noofrandomnums):
        randomnums[i] = (randomnums[i-1]*a +b) % c
        
 #Number of random numbers to be generated
noofrandomnums = 500

#To store random numbers
randomnums = [0] * noofrandomnums

#Function call
LCS(x0,a,b,c,randomnums,noofrandomnums)


index = []

#Print the generated random numbers
for i in range(len(randomnums)):
    if randomnums[i] == x0:    
        index.append(i)

print('Random Numbers =', randomnums)


print('period =', index[1] -index[0])
    
x = randomnums[1:]
#for i in range(len(randomnums)):
plt.plot(randomnums[:-1],x, '.')
plt.xlabel('$x_i$')
plt.ylabel('$x_(i+1)$')
plt.title('2D scatter plot of the pairs($x_i, x_{(i+1)}$)')
plt.show()

plt.plot(randomnums[:c], '.')
plt.title('2D scatter plot of the pairs ($x_i, x_{(i+1)}$) for one period')
plt.show()

    
```

In 1.c) again the Linear congruential random generator was used for a different sets of (a,b,c,$x_0$) values. For uniformity a kolmogorov-Smirnov test was used. From calculation we obtained that the test statistic was 0.977 and corresponding p-value was 0. Since the p-value is less that .05, we reject the null hypothesis, ans thus we can confirm that the random number generator doesn't follow normal distribution. Furthermore, autocorrelation at lag j was carried out in-order to reject the hypothesis that the samples drawn have any correlation as:
$$
\rho_{ij} = \frac{\sum_{k=0}^M \cdot x^{'}_{i + kj}\cdot x^{'}_{i (k + 1)j} }{\sum_{k=0}^{M}\cdot (x^{'2}_{i + kj})^{0.5} \cdot (x^{'2}_{i ( k + 1)j})^{0.5}}
$$
where, N is the total numbers of samples, j $\in${0,1,2,.., N-1}, i $\in$ {0,1,..,j-1} and $x^{'}_i$ = $x_i$ - $\bar{x}$.
Printing our $\rho$, we found that it converged to zero in the limit of N tends to infinity for any i, and j which proves that the numbers are truly random. Furthermore, "run test" is also tested and a 2D histogram is plotted for the results of run test. From the plotted results of the run test in histogram, it can be seen that the results are neither consistently increasing nor decreasing monotonically, which shows the random nature of our given input. 

```python
from scipy.stats import kstest
import random

a = 8  
b = 1    
c = 399
x0 = 2   


def LCS(x0,a,b,c,randomnums,noofrandomnums):
    randomnums[0] = x0
    for i in range(1, noofrandomnums):
        randomnums[i] = (randomnums[i-1]*a +b) % c
        
noofrandomnums = 500


randomnums = [0] * noofrandomnums


LCS(x0,a,b,c,randomnums,noofrandomnums)

index = []

#Print the generated random numbers
for i in range(len(randomnums)):
    if randomnums[i] == x0:    
        index.append(i)

print('period =', index[1] -index[0])
#print('Random Numbers =', randomnums)


#Autocorrelation at lag j

x = np.array(randomnums)
xmean = np.mean(x)
x_prime = [a - xmean for a in x]
N = len(randomnums)

def rho_func(i,j):
    sum1 = 0
    sum2 = 0
    sum3 = 0
    max_=0
    for s in range(N):
        if ((i+(s+1)*j)<= N-1):
            max_= s
    M = max_
    if M == 0:
        return 0
    for k in range(M):
        #print('here')
        sum1 += (x_prime[i+k*j] * x_prime[i+(k+1)*j])
        sum2 += (x_prime[i +k*j])**2
        sum3 += (x_prime[i+(k+1)*j])**2 
    return(sum1/(np.sqrt(sum2)*np.sqrt(sum3)))
      
rho = []    
for j in range(N-1):
    rho_j = []
    for i in range (j-1):
        rho_j.append(rho_func(i,j))
    rho.append(rho_j)
   # print(rho_j[-5:])
#print(rho)


#Run-Test

counter = 0
ncounter = 0
a = []

for i in range(len(x)-1):
    if x[i+1] > x[i]:
        counter = counter +1 
        if ncounter != 0:
            a.append(ncounter)
            ncounter = 0
    
    else:
        if counter != 0:
            a.append(counter)
            counter = 0
        ncounter = ncounter -1 
        
a = a[:42]     


#Kolmogorov-Smirnov test
ks = kstest(x, "norm")



print('value of ks = ',ks)
#print('value of rho =',rho)       
plt.hist(randomnums)
plt.xlabel('Random Numbers')
plt.ylabel('Period')
plt.title('2-D histogram of random numbers')
plt.show()

plt.hist(a)
plt.xlabel('length of runs')
plt.title('2-D histogram showing the results of run test')
plt.show()

        
```

### Random walks and first passage time

For a 1D random walk with 0.5 probability of each step, a random number is generated for large number of realisations of random walks starting form origin. For unit steps n, the histogram of displacement  is plotted below. Furthermore, the PDF ($x_n$) for various values of n is shown in a single graph, which resembles the gaussian distribution.

For each n, the mean and variance are computed and plotted as a funciton of n. It is observed that every time the code is executed, a different value of the plot in the case of mean is observed where as a linearly increasing plot is observed in the case of variance.

At the last part, the program in modified such that the random walk started at the origin and was terminated at the barriers ($x_0$ = 5,10,15,20,25). The total number of steps that the walker could take was limited to 10k steps. Also, it was observed that, doubling the barrier size almost doubled the average length of step needed to reach the particular barrier.


```python
p = 0.5

x = 0 
xx0 = [5*i for i in range(1,6)]
print(xx0)

l = 1

       
   
ev_x = []
ev_x_sq = []
var = []
N = [4,8,16,32,64,128,256]
for n in N:
    print("For N = ", n)
    final_posx=[]
    final_posx_sq=[]
    k=0
    no = 1000
    while (k<no):
        x = 0
        for i in range(n):
            rl = random.random()
            #print(rl)
            if rl <= p:
                x += l  
            else:
                x -= l             
        final_posx.append(x)
        final_posx_sq.append(x**2)
        k+=1    
        
    plt.hist(final_posx)
    plt.title('PDF(x_n)')
    plt.show()
    evx =  np.sum(final_posx)/len(final_posx)
    print('Expected values of x, <x> = ',evx)
    evx_sq = np.sum(final_posx_sq)/len(final_posx_sq)
    ev_x.append(evx)
    ev_x_sq.append(evx_sq)
    var.append(evx_sq - evx**2)    
    print('Variance of x, <del (x^2)> = ',evx_sq - evx**2)
    
    
sns.distplot(final_posx, hist = False, kde = True, kde_kws = {'shade': True, 'linewidth': 3}, label = n)
plt.title('PDF(x) for various values of n')
plt.show()


#Random walk which starts at 'O' and terminates at a barrier.
evx_ = []
for x0 in xx0:

    No_steps=[]
    final_posx1 = []
    k=0
    no = 1000
    while (k<no):
        x = 0
        n = 0
        while x<x0:
            rl = np.random.random()
            n+=1
            #print(rl)
            if rl <= p:
                x += l  
            else:
                x -= l
            if n == 10000:
                break

        final_posx1.append(x)
        if (n!= 10000):
            No_steps.append(n)
        k+=1    
    evx1 =  np.mean(No_steps)
    evx_.append(evx1)
        
    
    
    
        


  
    print('Average number of steps taken to reach barrier =',evx1)
  
    plt.hist(No_steps) # label = ' xx0={}'.format(xx0[i])
    plt.title('Histogram of the arrival time for {} steps in log-scale'.format(x0))
    plt.xscale('log')
    plt.show()


plt.plot(xx0,evx_)
plt.xlabel('Position of barrier')
plt.ylabel('Mean steps to reach the barrier')
plt.show()
 

plt.figure(7, figsize= (8,6))
plt.plot(N, ev_x)
plt.title('Mean of x', fontsize = 14)
plt.savefig('7.png')

plt.figure(8, figsize= (8,6))
plt.plot(N, var)
plt.title('Variance vs N', fontsize = 14)
plt.savefig('8.png')



```
